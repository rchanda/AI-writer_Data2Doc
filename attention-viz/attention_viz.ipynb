```python
import numpy as np
file = open('attentions.txt', 'r')
lines = file.readlines()
print(len(lines))
values = []
for line in lines:
    tokens = line.strip().split()
    row = []
    for token in tokens:
        if '~' not in token:
            row.append(float(token))
    if len(row) > 0:
        values.append(row)
```

```python
values = np.array(values)
```

```python
for row in values:
    print(len(row))
```

```python


timesteps = 10
def create_heatmap(data, cmap='Blues'):
    yticklabels = ['input_word']*len(data)
    xticklabels = summary[0:timesteps]
    plt.figure(figsize=(40,120))
    ax = sns.heatmap(data, cmap=cmap, cbar = True, xticklabels=xticklabels, yticklabels=yticklabels, annot_kws={"size": 15, "weight":"bold"})
    plt.title('')
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
```

```python
create_heatmap(values[:, 0:timesteps])
```
